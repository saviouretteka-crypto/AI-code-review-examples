```python

#  EXAMPLE 1 — Division / Empty Collection

print("\n" + "=" * 65)
print("  EXAMPLE 1: average()  —  Division on Empty Collection")
print("=" * 65)

# ── Buggy version ─────────────────────────────────────────────────────────────
def average_buggy(numbers):
    return sum(numbers) / len(numbers)

# ── Analysis ──────────────────────────────────────────────────────────────────
ANALYSIS_1 = """
BUGS FOUND
──────────
1. ZeroDivisionError when `numbers` is empty — len([]) == 0.
2. No type validation — passing strings or None raises a misleading TypeError.
3. Returns a raw float with no context; callers can't distinguish
   "average is 0.0" from "no data provided".

EDGE CASES
──────────
  average_buggy([])          → ZeroDivisionError
  average_buggy(None)        → TypeError: object of type 'NoneType' has no len()
  average_buggy(["a", "b"])  → TypeError inside sum()
  average_buggy([0, 0, 0])   → 0.0  (valid, but looks like an error return)
"""
print(ANALYSIS_1)

# ── Fixed version ─────────────────────────────────────────────────────────────
def average(numbers: list[float], default: float | None = None) -> float | None:
    """
    Return the arithmetic mean of a list of numbers.

    Args:
        numbers: List of numeric values.
        default: Value to return when the list is empty (default: None).

    Returns:
        Mean as a float, or `default` if the list is empty.

    Raises:
        TypeError: If `numbers` is not iterable or contains non-numeric values.
    """
    if not isinstance(numbers, (list, tuple)):
        raise TypeError(f"Expected list or tuple, got {type(numbers).__name__}")
    if not numbers:
        return default
    if not all(isinstance(n, (int, float)) for n in numbers):
        raise TypeError("All elements must be numeric.")
    return sum(numbers) / len(numbers)

# ── Tests ─────────────────────────────────────────────────────────────────────
def test_average():
    assert average([1, 2, 3])          == 2.0,   "basic mean failed"
    assert average([])                  is None,  "empty → None failed"
    assert average([], default=0.0)    == 0.0,   "empty with default failed"
    assert average([5])                == 5.0,   "single element failed"
    assert average([0, 0, 0])          == 0.0,   "all-zeros failed"
    assert average([-1, 1])            == 0.0,   "negatives failed"
    try:
        average(None)
        assert False, "should have raised TypeError"
    except TypeError:
        pass
    try:
        average(["a", "b"])
        assert False, "should have raised TypeError"
    except TypeError:
        pass
    print("  ✔  average() — all tests passed")

test_average()
```

```python
#  EXAMPLE 2 — Off-by-One / Range Error

print("\n" + "=" * 65)
print("  EXAMPLE 2: find_max_subarray_sum()  —  Off-by-One")
print("=" * 65)

# ── Buggy version ─────────────────────────────────────────────────────────────
def find_max_subarray_sum_buggy(arr):
    max_sum = 0
    current_sum = 0
    for num in arr:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
    return max_sum

# ── Analysis ──────────────────────────────────────────────────────────────────
ANALYSIS_2 = """
BUGS FOUND
──────────
1. Initialising max_sum = 0 means all-negative arrays incorrectly return 0
   instead of the least-negative element.
   Example: [-3, -1, -2] should return -1, but returns 0.

2. The algorithm itself (Kadane's) is correct for mixed arrays, but the
   initialisation breaks the all-negative contract.

EDGE CASES
──────────
  find_max_subarray_sum_buggy([-3, -1, -2])  → 0   (WRONG, should be -1)
  find_max_subarray_sum_buggy([-5])           → 0   (WRONG, should be -5)
  find_max_subarray_sum_buggy([])             → 0   (ambiguous — no data)
  find_max_subarray_sum_buggy([2, -1, 3])     → 4   (correct by luck)
"""
print(ANALYSIS_2)

# ── Fixed version ─────────────────────────────────────────────────────────────
def find_max_subarray_sum(arr: list[int | float]) -> int | float:
    """
    Return the maximum subarray sum using Kadane's algorithm.

    Correctly handles all-negative arrays by returning the largest
    (least-negative) single element.

    Args:
        arr: Non-empty list of numbers.

    Returns:
        Maximum contiguous subarray sum.

    Raises:
        ValueError: If arr is empty.
        TypeError:  If arr contains non-numeric values.
    """
    if not arr:
        raise ValueError("Array must not be empty.")
    if not all(isinstance(n, (int, float)) for n in arr):
        raise TypeError("All elements must be numeric.")

    # Seed with first element — handles all-negative arrays correctly
    max_sum     = arr[0]
    current_sum = arr[0]

    for num in arr[1:]:
        current_sum = max(num, current_sum + num)
        max_sum     = max(max_sum, current_sum)

    return max_sum

# ── Tests ─────────────────────────────────────────────────────────────────────
def test_find_max_subarray_sum():
    assert find_max_subarray_sum([1, -2, 3, 4, -1])  == 7,  "mixed failed"
    assert find_max_subarray_sum([-3, -1, -2])        == -1, "all-negative failed"
    assert find_max_subarray_sum([-5])                == -5, "single negative failed"
    assert find_max_subarray_sum([5])                 == 5,  "single positive failed"
    assert find_max_subarray_sum([2, -1, 2])          == 3,  "wrap-around failed"
    assert find_max_subarray_sum([1, 2, 3, 4])        == 10, "all-positive failed"
    try:
        find_max_subarray_sum([])
        assert False, "should have raised ValueError"
    except ValueError:
        pass
    print("  ✔  find_max_subarray_sum() — all tests passed")

test_find_max_subarray_sum()
```

```python
#  EXAMPLE 3 — Mutable Default Argument

print("\n" + "=" * 65)
print("  EXAMPLE 3: add_item()  —  Mutable Default Argument")
print("=" * 65)

# ── Buggy version ─────────────────────────────────────────────────────────────
def add_item_buggy(item, items=[]):
    items.append(item)
    return items

# ── Analysis ──────────────────────────────────────────────────────────────────
ANALYSIS_3 = """
BUGS FOUND
──────────
1. The default argument `items=[]` is evaluated ONCE at function definition
   time, not on every call. All calls that omit `items` share the same list.
   This causes state to leak between calls — a classic Python gotcha.

2. The function mutates its input in-place AND returns it, making the caller's
   reference and the return value the same object. Callers may not expect this.

EDGE CASES
──────────
  a = add_item_buggy("x")   → ["x"]          (looks fine)
  b = add_item_buggy("y")   → ["x", "y"]     (WRONG — should be ["y"])
  a is b                    → True            (same list object!)
  add_item_buggy("z", [])   → ["z"]           (fine — explicit list bypasses bug)
"""
print(ANALYSIS_3)

# ── Fixed version ─────────────────────────────────────────────────────────────
def add_item(item: object, items: list | None = None) -> list:
    """
    Return a new list with `item` appended.

    Uses None as the sentinel default to avoid mutable default argument
    pitfall. Always returns a new list; never mutates the input.

    Args:
        item:  Any value to append.
        items: Existing list to extend (optional). If omitted, starts fresh.

    Returns:
        A new list containing all original items plus the new one.
    """
    if items is None:
        items = []
    return [*items, item]   # spread into new list — never mutates input

# ── Tests ─────────────────────────────────────────────────────────────────────
def test_add_item():
    a = add_item("x")
    b = add_item("y")
    assert a == ["x"],    "first call contaminated"
    assert b == ["y"],    "second call contaminated"
    assert a is not b,    "calls share the same list object"

    original = [1, 2]
    result   = add_item(3, original)
    assert result   == [1, 2, 3], "extend failed"
    assert original == [1, 2],    "input was mutated"
    print("  ✔  add_item() — all tests passed")

test_add_item()
```

```python
#  EXAMPLE 4 — Silent Data Loss in String Processing

print("\n" + "=" * 65)
print("  EXAMPLE 4: count_words()  —  Silent Data Loss")
print("=" * 65)

# ── Buggy version ─────────────────────────────────────────────────────────────
def count_words_buggy(text):
    words = text.split(" ")
    return len(words)

# ── Analysis ──────────────────────────────────────────────────────────────────
ANALYSIS_4 = """
BUGS FOUND
──────────
1. split(" ") splits on single spaces only. Multiple consecutive spaces
   produce empty strings ("") in the result, inflating the count.
   "hello  world".split(" ") → ["hello", "", "world"] → len = 3, not 2.

2. Leading/trailing whitespace creates phantom empty-string tokens.
   "  hello  ".split(" ") → ["", "", "hello", "", ""] → len = 5, not 1.

3. No handling for non-string input (None, int, etc.).

4. Tabs, newlines, and other whitespace are not treated as word separators.

EDGE CASES
──────────
  count_words_buggy("hello  world")   → 3   (WRONG, should be 2)
  count_words_buggy("  hi  ")         → 5   (WRONG, should be 1)
  count_words_buggy("")               → 1   (WRONG, should be 0)
  count_words_buggy("hello\\tworld")   → 1   (WRONG, tab not split)
  count_words_buggy(None)             → AttributeError
"""
print(ANALYSIS_4)

# ── Fixed version ─────────────────────────────────────────────────────────────
def count_words(text: str) -> int:
    """
    Count the number of words in a string.

    Uses str.split() with no argument, which splits on any whitespace
    (spaces, tabs, newlines) and automatically ignores leading, trailing,
    and consecutive whitespace — returning [] for blank/empty strings.

    Args:
        text: Input string to count words in.

    Returns:
        Number of words (0 for empty or whitespace-only strings).

    Raises:
        TypeError: If `text` is not a string.
    """
    if not isinstance(text, str):
        raise TypeError(f"Expected str, got {type(text).__name__}")
    return len(text.split())   # no arg → splits on any whitespace, skips empties

# ── Tests ─────────────────────────────────────────────────────────────────────
def test_count_words():
    assert count_words("hello world")      == 2, "basic failed"
    assert count_words("hello  world")     == 2, "double-space failed"
    assert count_words("  hi  ")           == 1, "leading/trailing space failed"
    assert count_words("")                 == 0, "empty string failed"
    assert count_words("   ")              == 0, "whitespace-only failed"
    assert count_words("hello\tworld")     == 2, "tab separator failed"
    assert count_words("a\nb\nc")          == 3, "newline separator failed"
    assert count_words("one")              == 1, "single word failed"
    try:
        count_words(None)
        assert False, "should have raised TypeError"
    except TypeError:
        pass
    print("  ✔  count_words() — all tests passed")

test_count_words()
```

```python
#  EXAMPLE 5 — Incorrect Loop Mutation

print("\n" + "=" * 65)
print("  EXAMPLE 5: remove_negatives()  —  Mutating While Iterating")
print("=" * 65)

# ── Buggy version ─────────────────────────────────────────────────────────────
def remove_negatives_buggy(numbers):
    for num in numbers:
        if num < 0:
            numbers.remove(num)
    return numbers

# ── Analysis ──────────────────────────────────────────────────────────────────
ANALYSIS_5 = """
BUGS FOUND
──────────
1. Modifying a list while iterating over it causes the iterator to skip
   elements. When an item is removed, all subsequent items shift left by one
   index, but the iterator advances its internal pointer — skipping the next
   element.

2. list.remove() removes the FIRST occurrence of a value, not the element
   at the current index. Two adjacent equal negatives will only remove one.

3. The function mutates the caller's list in-place (no copy), which is
   unexpected and dangerous when the caller still holds a reference.

EDGE CASES
──────────
  remove_negatives_buggy([-1, -2, 3])   → [-2, 3]   (WRONG, -2 skipped)
  remove_negatives_buggy([-1, -1, 1])   → [-1, 1]   (WRONG, duplicate missed)
  remove_negatives_buggy([-3, -3, -3])  → [-3]       (WRONG, two survive)
"""
print(ANALYSIS_5)

# ── Fixed version ─────────────────────────────────────────────────────────────
def remove_negatives(numbers: list[int | float]) -> list[int | float]:
    """
    Return a new list with all negative numbers removed.

    Never mutates the input list. Uses a list comprehension for clarity
    and correctness — no iterator invalidation risk.

    Args:
        numbers: List of numeric values.

    Returns:
        New list containing only non-negative values (>= 0).

    Raises:
        TypeError: If input is not a list or contains non-numeric values.
    """
    if not isinstance(numbers, list):
        raise TypeError(f"Expected list, got {type(numbers).__name__}")
    if not all(isinstance(n, (int, float)) for n in numbers):
        raise TypeError("All elements must be numeric.")
    return [n for n in numbers if n >= 0]

# ── Tests ─────────────────────────────────────────────────────────────────────
def test_remove_negatives():
    assert remove_negatives([-1, -2, 3])   == [3],    "basic removal failed"
    assert remove_negatives([-1, -1, 1])   == [1],    "duplicate negatives failed"
    assert remove_negatives([-3, -3, -3])  == [],     "all-negative failed"
    assert remove_negatives([1, 2, 3])     == [1,2,3],"no negatives failed"
    assert remove_negatives([0, -1, 0])    == [0, 0], "zero preserved failed"
    assert remove_negatives([])            == [],     "empty list failed"

    # Verify input not mutated
    original = [-1, 2, -3]
    _        = remove_negatives(original)
    assert original == [-1, 2, -3], "input was mutated"
    print("  ✔  remove_negatives() — all tests passed")

test_remove_negatives()



#  EXAMPLE 6 — Race Condition in File Handling

print("\n" + "=" * 65)
print("  EXAMPLE 6: read_config()  —  TOCTOU Race Condition")
print("=" * 65)

import json
import os

# ── Buggy version ─────────────────────────────────────────────────────────────
def read_config_buggy(filepath):
    if os.path.exists(filepath):       # CHECK
        with open(filepath) as f:      # USE  ← gap between check and use
            return json.load(f)
    return {}

# ── Analysis ──────────────────────────────────────────────────────────────────
ANALYSIS_6 = """
BUGS FOUND
──────────
1. TOCTOU (Time-of-Check / Time-of-Use) race condition: the file could be
   deleted, renamed, or have its permissions changed between the os.path.exists()
   check and the open() call — causing an unhandled exception anyway.

2. No error handling for malformed JSON — json.JSONDecodeError will propagate
   uncaught, crashing the caller with an unhelpful traceback.

3. No handling for PermissionError (file exists but is unreadable).

4. Silently returns {} on missing file — callers can't distinguish
   "file not found" from "file was empty / config is intentionally empty".

EDGE CASES
──────────
  File deleted between check and open  → FileNotFoundError (race)
  File contains invalid JSON           → json.JSONDecodeError (uncaught)
  File exists but no read permission   → PermissionError (uncaught)
  File is empty                        → json.JSONDecodeError (uncaught)
"""
print(ANALYSIS_6)

# ── Fixed version ─────────────────────────────────────────────────────────────
from pathlib import Path

class ConfigError(Exception):
    """Raised when a config file cannot be read or parsed."""

def read_config(filepath: str | Path, default: dict | None = None) -> dict:
    """
    Read and parse a JSON config file.

    Eliminates the TOCTOU race by attempting the open() directly and
    handling exceptions explicitly. Never silently swallows errors that
    would hide bugs.

    Args:
        filepath: Path to the JSON config file.
        default:  Value to return if the file does not exist.
                  Pass None (default) to raise on missing file instead.

    Returns:
        Parsed config dict.

    Raises:
        ConfigError: If the file is missing (and no default given),
                     unreadable, or contains invalid JSON.
    """
    path = Path(filepath)
    try:
        with open(path, encoding="utf-8") as f:
            content = f.read().strip()
            if not content:
                return default if default is not None else {}
            return json.loads(content)

    except FileNotFoundError:
        if default is not None:
            return default
        raise ConfigError(f"Config file not found: {path}") from None

    except PermissionError:
        raise ConfigError(f"Permission denied reading: {path}") from None

    except json.JSONDecodeError as e:
        raise ConfigError(f"Invalid JSON in {path}: {e}") from e

# ── Tests ─────────────────────────────────────────────────────────────────────
import tempfile

def test_read_config():
    # Valid JSON file
    with tempfile.NamedTemporaryFile("w", suffix=".json", delete=False) as f:
        json.dump({"key": "value"}, f)
        tmp = f.name
    assert read_config(tmp) == {"key": "value"}, "valid JSON failed"
    os.unlink(tmp)

    # Missing file with default
    assert read_config("/no/such/file.json", default={}) == {}, "missing + default failed"

    # Missing file without default → ConfigError
    try:
        read_config("/no/such/file.json")
        assert False, "should have raised ConfigError"
    except ConfigError:
        pass

    # Invalid JSON
    with tempfile.NamedTemporaryFile("w", suffix=".json", delete=False) as f:
        f.write("not json {{")
        tmp = f.name
    try:
        read_config(tmp)
        assert False, "should have raised ConfigError"
    except ConfigError:
        pass
    finally:
        os.unlink(tmp)

    print("  ✔  read_config() — all tests passed")

test_read_config()
```
